```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
def load_data():
    with open('X_train.npy', 'rb') as f:
        X_train = np.load(f)
    with open('Y_train.npy', 'rb') as f:
        Y_train = np.load(f)
    with open('X_test.npy', 'rb') as f:
        X_test = np.load(f)
    with open('Y_test.npy', 'rb') as f:
        Y_test = np.load(f)
    f.close()
    return X_train,X_test,Y_train,Y_test
```

```python
def initalise(dems):
    W = np.zeros((dems,1))
    b = 0.0 
    return W,b
```

```python
def sigmoid(z):
    s = 1/(1+np.exp(-z))
    return s
```

```python
def logistic_regression(W,b,m,learning_rate,X_train,Y_train):
    Z = np.dot(W.T,X_train) + b
    A = sigmoid(Z)
    dZ = A - Y_train
    dW = np.dot(X_train,dZ.T)/m
    db = np.sum(dZ)/m
    W = W - learning_rate*dW
    b = b - learning_rate*db
    return W,b
```

```python
def accuracy(W,b,m_test,X_test,Y_test):
    Z = np.dot(W.T,X_test) + b
    Y_out = sigmoid(Z)
    count = 0
    for i in range (m_test):
        if Y_out[0][i]<=0.5:
            out = 0
        else:
            out = 1
        if out == Y_test[0][i]:
            count = count + 1
    acc = (count/m_test)*100
    return acc
```

```python
X_train,X_test,Y_train,Y_test = load_data()
W,b = initalise(8*8*3)
m = 25000
m_test = 2000
epoch = 1000
learning_rate = 0.001
for i in range (epoch):
    W,b=logistic_regression(W,b,m,learning_rate,X_train,Y_train)
    acc = accuracy(W,b,m_test,X_test,Y_test)
    print(str(acc)+"%")
```
