```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
from PIL import Image 
import os 
import random
```

```python
def load_processing(image):
    image_load = Image.open(image)
    image_load = image_load.resize((8,8))
    arr = np.array(image_load)
    arr = arr.reshape(8*8*3,1)
    return arr
```

```python
#indices = list(range(0,25000))
indices = list(range(1,2001))
random.shuffle(indices)
```

```python
i = 0
indice = indices[i]
if(indice<=1000):
    image = "./dataset/dogs-vs-cats/test/cats/cat."+str(indice)+".jpg"
else:
    image = "./dataset/dogs-vs-cats/test/dogs/dog."+str(indice-1000)+".jpg"
arr = load_processing(image)
X = np.array(arr).reshape(len(arr),1)
Y = np.array([1])
idx = 1
for i in range(1,2000):
    indice = indices[i]
    if indice<=1000:
        image = "./dataset/dogs-vs-cats/test/cats/cat."+str(indice)+".jpg"
        arr = load_processing(image)
        X = np.hstack((X,arr))
        Y = np.append(Y,[1])
    else:
        image = "./dataset/dogs-vs-cats/test/dogs/dog."+str(indice-1000)+".jpg"
        arr = load_processing(image)
        X = np.hstack((X,arr))
        Y = np.append(Y,[0])
    idx = idx + 1
    print(str(idx/20)+"%")
```
